```python
# read train.txt, valid.txt, test.txt and replace each id with the corresponding name in the ind2ent.pkl and ind2rel.pkl which are dictionaries
import pickle

with open('ind2ent.pkl', 'rb') as f:
    ind2ent = pickle.load(f)
with open('ind2rel.pkl', 'rb') as f:
    ind2rel = pickle.load(f)
    
def replace_ids_with_names(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    new_lines = []
    for line in lines:
        head, relation, tail = line.strip().split('\t')
        head_name = ind2ent[int(head)]
        relation_name = ind2rel[int(relation)]
        tail_name = ind2ent[int(tail)]
        new_lines.append(f"{head_name}\t{relation_name}\t{tail_name}\n")
    # file_path is like test-id.txt, we the new file should be test.txt
    new_file_path = file_path.replace('-id', '')
    with open(new_file_path, 'w') as f:
        f.writelines(new_lines)
replace_ids_with_names('train-id.txt')
replace_ids_with_names('valid-id.txt')
replace_ids_with_names('test-id.txt')
```
